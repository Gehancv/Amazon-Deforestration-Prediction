```python
#Read deforestration dataset
df = read_csv('formatted/terrabrasilis_deforestration.csv', decimal=',')
#Convert data type of area to float
df['area'] = df['area'].str.replace(',','')
df['area'] = df.area.astype(pd.np.float64)

#Print shape of the dataset
print('Shape: {}'.format(df.shape))

#Print first 10 rows of the dataset
df.head(10)
```

```python
#Finding missing values
cols = df.columns
print(df[cols].isnull().sum())

#Describe statistics of each column in the dataset
df.describe(include='all')
```

There are deforestration information related to 23 states(i.e 'uf') and 5 regions. <br>

```python
#Total area lost annualy in each forest region
regions = df.region.unique()
years = df.year.unique()
area_by_region = pd.DataFrame(columns=['region', 'year', 'area'])
i=0
for reg in regions:
    for yr in years:
        res = round(df.loc[(df['region']==reg) & (df['year']==yr), 'area'].sum(), 3)
        area_by_region.loc[i,:] = [reg, yr,res]
        i = i+1
area_by_region['year'] = area_by_region.year.astype(pd.np.float64)
area_by_region['area'] = area_by_region.area.astype(pd.np.float64)

#Print shape of the dataset
print('Shape: {}'.format(area_by_region.shape))

#Print first 10 rows of the dataset
area_by_region.head(10)
```

```python
#Total area lost annualy in each forest region
states = df.uf.unique()
years = df.year.unique()
area_by_state = pd.DataFrame(columns=['state', 'year', 'area'])
i=0
for state in states:
    for yr in years:
        res = round(df.loc[(df['uf']==state) & (df['year']==yr), 'area'].sum(), 3)
        area_by_state.loc[i,:] = [state, yr,res]
        i = i+1
area_by_state['year'] = area_by_state.year.astype(pd.np.float64)
area_by_state['area'] = area_by_state.area.astype(pd.np.float64)

#Print shape of the dataset
print('Shape: {}'.format(area_by_state.shape))

#Print first 10 rows of the dataset
area_by_state.head(10)
```

```python
#Plot total area of deforestration over the years by forest region
import matplotlib.pyplot as plt
%matplotlib inline

line_1 = area_by_region.loc[area_by_region.region == 'pantanal'].sort_values(by='year')
line_2 = area_by_region.loc[area_by_region.region == 'pampa'].sort_values(by='year')
line_3 = area_by_region.loc[area_by_region.region == 'cerrado'].sort_values(by='year')
line_4 = area_by_region.loc[area_by_region.region == 'mata_atlantica'].sort_values(by='year')
line_5 = area_by_region.loc[area_by_region.region == 'caatinga'].sort_values(by='year')

fig,ax = plt.subplots()

ax.plot(line_1.year, line_1['area'], color='b')
ax.plot(line_2.year, line_2['area'], color='g')
ax.plot(line_3.year, line_3['area'], color='r')
ax.plot(line_4.year, line_4['area'], color='y')
ax.plot(line_5.year, line_5['area'], color='c')
ax.legend(labels=['pantanal','pampa','cerrado','mata_atlantica','caatinga'])
plt.show()
```

Mata Atlantica forest region has the highest deforestration by land area.
